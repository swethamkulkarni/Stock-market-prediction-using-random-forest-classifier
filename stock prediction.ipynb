```python
import yfinance as yf# yfinance as a yahoo finance api 
```

```python
ticker= yf.Ticker("^GSPC")#ticker is a constructor in yfinance lib
```

```python
sp500= ticker.history(period="max")# max wil give u all data available, 1d wil give 1day data, and 1mo will give one month data
```

```python
sp500=sp500.drop(['Dividends','Stock Splits'], axis=1).copy()# or del sp500['Dividends']
```

```python
sp500
```

```python
#target here will be to predict either stock will go up or down
sp500['Tomorrow']= sp500['Close'].shift(-1)
```

```python
sp500
```

```python
sp500['Target']= (sp500['Tomorrow']>sp500['Close']).astype(int)#astype is used in pandas to convert the data types
```

```python
sp500
```

```python
sp500=sp500.loc["1990-01-01":].copy()
```

```python
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier

model= RandomForestClassifier(n_estimators=100, min_samples_split=100,class_weight='balanced',random_state=1)
train = sp500.iloc[:-500]
test=sp500.iloc[-500:]

predictors=["Close","Volume","Open","High","Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score
preds= model.predict(test[predictors])
```

```python
from sklearn.metrics import classification_report
print(classification_report(test["Target"],preds))
```

```python
import pandas as pd
preds=pd.Series(preds, index= test.index)
```

```python
preds
```

```python
precision_score(test['Target'],preds,zero_division=0)
```

```python
combined= pd.concat([test["Target"],preds],axis=1)
```

```python
#backtesting
def predict(train,test, predictors,model):
    model.fit(train[predictors],train["Target"])
    preds=model.predict(test[predictors])
    preds=pd.Series(preds,index=test.index,name="Predictions")
    combined=pd.concat([test["Target"],preds],axis=1)
    return combined
```

```python
def backtest(data, model,predictors, start=2500,step=250):
    all_predictions=[]
    for i in range(start,data.shape[0],step):
        train=data.iloc[0:i].copy()
        test=data.iloc[i:(i+step)].copy()
        predictions=predict(train,test,predictors,model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500,model,predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"],predictions["Predictions"])
```

```python
#percentage of days that market actually went up
predictions["Target"].value_counts()/predictions.shape[0]
```

```python
#these are number of days we will look to decide if the stock went up and then we wil find ratio between todays closing value to these days closing value and if its up then there might be down swing and vice-versa
horizons=[2,5,60,250,1000]
new_predictors=[]
for horizon in horizons:
    rolling_averages=sp500.rolling(horizon).mean()
    ratio_column=f"Close_Ratio_{horizon}"
    sp500[ratio_column]=sp500["Close"]/rolling_averages["Close"]
    
    #trend here is number of days in past X days whatever the horizon is that the stock price actually went up
    trend_column=f" Trend{horizon}"
    sp500[trend_column]= sp500.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors += [ratio_column,trend_column]
```

```python
from sklearn.experimental import enable_hist_gradient_boosting  # noqa
from sklearn.ensemble import HistGradientBoostingClassifier

# Define the model
model = HistGradientBoostingClassifier()

```

```python
def predict(train,test, predictors,model):
    model.fit(train[predictors],train["Target"])
    preds=model.predict_proba(test[predictors])[:,1]#returns the probability if the market goes up or down
    preds[preds >= .6] = 1 
    preds[preds < .6] = 0
    preds=pd.Series(preds,index=test.index,name="Predictions")
    combined=pd.concat([test["Target"],preds],axis=1)
    return combined
```

```python
predictions= backtest(sp500,model,new_predictors)
```

```python
precision_score(predictions["Target"],predictions["Predictions"])
```
